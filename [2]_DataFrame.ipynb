### 1. Narysuj wykres zależności kolumn Birthrate i Deathrate w zależności od GDP per capita dla 30 najbogatszych i 30 najbiedniejszych krajów.

```python
countries_df = spark.read.csv("countries of the world.csv", header=True, inferSchema=True)
```

```python
# Sortowanie według GDP ($ per capita)
sorted_countries_df = countries_df.orderBy(countries_df["GDP ($ per capita)"].desc())
```

```python
# Wybór 30 najbogatszych i 30 najbiedniejszych krajów
top_30_richest = sorted_countries_df.limit(30)
bottom_30_poorest = sorted_countries_df.orderBy(countries_df["GDP ($ per capita)"].asc()).limit(30)
```

```python
rich_pandas_df = top_30_richest.toPandas()
poor_pandas_df = bottom_30_poorest.toPandas()
```

```python
# Zamiana przecinków na kropki w kolumnach Birthrate i Deathrate
rich_pandas_df['Birthrate'] = rich_pandas_df['Birthrate'].str.replace(',', '.')
rich_pandas_df['Deathrate'] = rich_pandas_df['Deathrate'].str.replace(',', '.')
poor_pandas_df['Birthrate'] = poor_pandas_df['Birthrate'].str.replace(',', '.')
poor_pandas_df['Deathrate'] = poor_pandas_df['Deathrate'].str.replace(',', '.')
```

```python
# Konwersja kolumn na liczby zmiennoprzecinkowe
rich_pandas_df['Birthrate'] = pd.to_numeric(rich_pandas_df['Birthrate'], errors='coerce')
rich_pandas_df['Deathrate'] = pd.to_numeric(rich_pandas_df['Deathrate'], errors='coerce')

poor_pandas_df['Birthrate'] = pd.to_numeric(poor_pandas_df['Birthrate'], errors='coerce')
poor_pandas_df['Deathrate'] = pd.to_numeric(poor_pandas_df['Deathrate'], errors='coerce')
```

```python
# Usuwanie wierszy z brakującymi wartościami
rich_pandas_df = rich_pandas_df.dropna(subset=['Birthrate', 'Deathrate', 'GDP ($ per capita)'])
poor_pandas_df = poor_pandas_df.dropna(subset=['Birthrate', 'Deathrate', 'GDP ($ per capita)'])
```

```python
plt.figure(figsize=(10, 6))

# Wykres dla 30 najbogatszych krajów
plt.scatter(rich_pandas_df['GDP ($ per capita)'], rich_pandas_df['Birthrate'], color='green', label='Richest 30 Birthrate')
plt.scatter(rich_pandas_df['GDP ($ per capita)'], rich_pandas_df['Deathrate'], color='blue', label='Richest 30 Deathrate')

# Wykres dla 30 najbiedniejszych krajów
plt.scatter(poor_pandas_df['GDP ($ per capita)'], poor_pandas_df['Birthrate'], color='red', label='Poorest 30 Birthrate')
plt.scatter(poor_pandas_df['GDP ($ per capita)'], poor_pandas_df['Deathrate'], color='purple', label='Poorest 30 Deathrate')

plt.xlabel('GDP ($ per capita)')
plt.ylabel('Birthrate / Deathrate')
plt.legend()
plt.show()
```

### 4.Na wykresie scatterplot narysuj zależność liczby lotnisk od powierzchni kraju.


#### * Dane potrzebne do wykresu są w dwóch ramkach (Countries i Airports), konieczne więc będzie ich złączenie (join).
#### * Nie zawsze nazwy tych samych krajów będą identyczne w obu ramkach, co może stanowić problem (złączenie nie powstanie). Zastanów się jak przy pomocy mechanizmu złączeń wykryć, które nazwy się różnią. Stwórz odpowiednią ramkę, która zawiera takie nazwy.

```python
# Usunięcie spacji z nazw krajów
countries_df = countries_df.withColumn('Country', trim(countries_df['Country']))
airports_df = airports_df.withColumn('Country', trim(airports_df['Country']))

# Pobranie unikalnych krajów z countries_df i airports_df
countries_unique = set(countries_df.select('Country').distinct().rdd.flatMap(lambda x: x).collect())
airports_unique = set(airports_df.select('Country').distinct().rdd.flatMap(lambda x: x).collect())

# Znalezienie różnic między unikalnymi nazwami krajów
country_differences = countries_unique.symmetric_difference(airports_unique)
print("Różnice w nazwach krajów:", country_differences)
```

```python
# Stworzenie mapowania nazw krajów 
mapping_expr = when(col('Country') == 'Antigua & Barbuda', 'Antigua and Barbuda') \
    .when(col('Country') == 'Bahamas, The', 'Bahamas') \
    .when(col('Country') == 'Bosnia & Herzegovina', 'Bosnia and Herzegovina') \
    .when(col('Country') == 'British Virgin Is.', 'British Virgin Islands') \
    .when(col('Country') == 'Central African Rep.', 'Central African Republic') \
    .when(col('Country') == 'Congo, Repub. of the', 'Congo (Brazzaville)') \
    .when(col('Country') == 'Congo, Dem. Rep.', 'Congo (Kinshasa)') \
    .when(col('Country') == 'Gambia, The', 'Gambia') \
    .when(col('Country') == 'Korea, North', 'North Korea') \
    .when(col('Country') == 'Korea, South', 'South Korea') \
    .when(col('Country') == 'Micronesia, Fed. St.', 'Micronesia') \
    .when(col('Country') == 'N. Mariana Islands', 'Northern Mariana Islands') \
    .when(col('Country') == 'Saint Kitts & Nevis', 'Saint Kitts and Nevis') \
    .when(col('Country') == 'St Pierre & Miquelon', 'Saint Pierre and Miquelon') \
    .when(col('Country') == 'Sao Tome & Principe', 'Sao Tome and Principe') \
    .when(col('Country') == 'Trinidad & Tobago', 'Trinidad and Tobago') \
    .when(col('Country') == 'Turks & Caicos Is', 'Turks and Caicos Islands') \
    .otherwise(col('Country'))

# Stosowanie mapowania
countries_df = countries_df.withColumn('Country', mapping_expr)
```

```python
# Zmiana nazwy kolumny "Area (sq. mi.)" na Area_sq_mi
countries_df = countries_df.withColumnRenamed('Area (sq. mi.)', 'Area_sq_mi')
```

```python
# Zapis ramek danych jako widoków tymczasowych dla SQL
countries_df.createOrReplaceTempView("countries")
airports_df.createOrReplaceTempView("airports")
```

```python
# Znalezienie różnic w nazwach krajów
differences_query = """
SELECT
    COALESCE(c.Country, a.Country) AS Country,
    c.Area_sq_mi,
    a.`Airport ID`
FROM
    countries c
FULL OUTER JOIN
    airports a
ON
    c.Country = a.Country
WHERE
    c.Area_sq_mi IS NULL OR a.`Airport ID` IS NULL
"""

differences_df = spark.sql(differences_query)
```

```python
# Liczba lotnisk na kraj
airports_count_by_country = airports_df.groupBy('Country').agg(count('Airport ID').alias('Airport_Count'))
```

```python
# Wynik jako widok tymczasowy
airports_count_by_country.createOrReplaceTempView("airports_count")
```

```python
# Połączenie danych o powierzchni kraju i liczbie lotnisk
merge_query = """
SELECT
    c.Country,
    c.Area_sq_mi,
    a.Airport_Count
FROM
    countries c
INNER JOIN
    airports_count a
ON
    c.Country = a.Country
"""
merged_df = spark.sql(merge_query)
```

```python
plot_data = merged_df.toPandas()
plot_data.plot(x='Area_sq_mi', y='Airport_Count', kind='scatter')

plt.xlabel('Surface Area (sq mi)')
plt.ylabel('Number of Airports')
plt.title('Number of Airports vs Surface Area by Country')

plt.show()
```

### 5.Oblicz, ile lotnisk jest na poszczególnych kontynentach. To zadanie również wymaga złączenia ramek "Countries" i "Airports". Wynik przedstaw na wykresie kołowym.

```python
# Mapowanie krajów na kontynenty
continent_mapping = {
    'Europe': ['Andorra', 'Austria', 'Belgium', 'Bosnia and Herzegovina', 'Croatia', 'Denmark', 'France', 'Germany', 'Greece', 'Italy', 'Liechtenstein', 'Monaco', 'Montenegro', 'Netherlands', 'Poland', 'Portugal', 'Spain', 'Switzerland', 'United Kingdom'],
    'Asia': ['Afghanistan', 'China', 'India', 'Indonesia', 'Japan', 'Malaysia', 'Nepal', 'Pakistan', 'Singapore', 'South Korea', 'Thailand', 'Vietnam'],
    'Africa': ['Algeria', 'Egypt', 'Nigeria', 'South Africa', 'Morocco', 'Kenya', 'Tunisia'],
    'North America': ['Canada', 'United States', 'Mexico', 'Bahamas', 'Cuba'],
    'South America': ['Argentina', 'Brazil', 'Chile', 'Colombia', 'Peru', 'Venezuela'],
    'Oceania': ['Australia', 'New Zealand', 'Fiji', 'Papua New Guinea'],
    'Antarctica': ['Antarctica']
}

# Dodanie kolumny kontynentu 
countries_df = countries_df.withColumn('Continent', 
    when(col('Country').isin(continent_mapping['Europe']), 'Europe')
    .when(col('Country').isin(continent_mapping['Asia']), 'Asia')
    .when(col('Country').isin(continent_mapping['Africa']), 'Africa')
    .when(col('Country').isin(continent_mapping['North America']), 'North America')
    .when(col('Country').isin(continent_mapping['South America']), 'South America')
    .when(col('Country').isin(continent_mapping['Oceania']), 'Oceania')
    .when(col('Country').isin(continent_mapping['Antarctica']), 'Antarctica')
    .otherwise(lit('Unknown'))
)
```

```python
# Zapis ramek danych jako widoków tymczasowych dla SQL
countries_df.createOrReplaceTempView("countries")
airports_df.createOrReplaceTempView("airports")
```

```python
# Połączenie danych o krajach i lotniskach na podstawie kolumny "Country"
merge_query = """
SELECT
    c.Continent,
    a.`Airport ID`
FROM
    countries c
INNER JOIN
    airports a
ON
    c.Country = a.Country
"""
merged_df = spark.sql(merge_query)
```

```python
# Grupowanie według kontynentu i zliczanie liczby lotnisk na każdy kontynent
airports_by_continent = merged_df.groupBy('Continent').agg(count('Airport ID').alias('Airport_Count'))
```

```python
plot_data = airports_by_continent.toPandas()

labels = plot_data['Continent']
sizes = plot_data['Airport_Count']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Number of airports on continents')
plt.axis('equal')  
plt.show()
```
